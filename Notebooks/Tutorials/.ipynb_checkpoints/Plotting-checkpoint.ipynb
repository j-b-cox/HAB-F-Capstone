```python
from cartopy.mpl.ticker import LongitudeFormatter, LatitudeFormatter
from IPython.display import JSON
import cartopy
import cartopy.crs as ccrs
import cartopy.feature as cfeature
import earthaccess
import matplotlib.pyplot as plt
import numpy as np
import xarray as xr
import pyresample
from pyresample import geometry, kd_tree

# ---------------------------
# Authenticate and Configure
# ---------------------------
auth = earthaccess.login(persist=True)

tspan = ("2020-08-15", "2020-08-23")
bbox = (-83.5, 41.3, -82.3, 42.2)
min_lat, max_lat = bbox[1], bbox[3]
min_lon, max_lon = bbox[0], bbox[2]

# ---------------------------
# Search and Download
# ---------------------------
results = earthaccess.search_data(
    short_name="MODISA_L2_OC",
    temporal=tspan,
    bounding_box=bbox,
    cloud_cover=(0, 50),
)

print("len(results):", len(results))
paths = earthaccess.download(results, "data")
print("len(paths):", len(paths))

# ---------------------------
# Load Data
# ---------------------------
for path in range(len(paths)):
    skip_file = False
    
    prod = xr.open_dataset(paths[0])
    obs = xr.open_dataset(paths[0], group="geophysical_data")
    nav = xr.open_dataset(paths[0], group="navigation_data")
    
    nav = nav.set_coords(("longitude", "latitude"))
    dataset = xr.merge((prod, obs, nav.coords))
    
    # ---------------------------
    # Define Target Grid
    # ---------------------------
    target_lons = np.arange(-180, 180, 0.01)
    target_lats = np.arange(-90, 90, 0.01)
    lon2d, lat2d = np.meshgrid(target_lons, target_lats)
    area_def = geometry.GridDefinition(lons=lon2d, lats=lat2d)
    
    # ---------------------------
    # Process 3 Bands for RGB
    # ---------------------------
    #bands = {
    #    "Rrs_678": None,  # Red
    #    "Rrs_555": None,  # Green
    #    "Rrs_488": None,  # Blue
    #}
    bands = {
        "Rrs_667": None,  # Red
        "Rrs_555": None,  # Green
        "Rrs_488": None,  # Blue
    }
    
    regridded_channels = []
    
    for band_name in bands:
        array = dataset[band_name]
    
        # Flatten lat/lon/data
        lons = nav["longitude"].values.flatten()
        lats = nav["latitude"].values.flatten()

        data = dataset[band_name].values.flatten()
    
        valid_mask = ~np.isnan(data) & ~np.isnan(lons) & ~np.isnan(lats)
        
        if np.mean(valid_mask) == 0:
            print("No valid data found in this file, skipping.")
            skip_file = True
    
        lons = lons[valid_mask]
        lats = lats[valid_mask]
        data = data[valid_mask]
        
        swath_def = geometry.SwathDefinition(lons=lons, lats=lats)
        
        result = kd_tree.resample_nearest(
            swath_def, data, area_def,
            radius_of_influence=5000,
            fill_value=np.nan
        )
        print(f"{band_name} valid pixel % (before clipping):", (np.count_nonzero(~np.isnan(result)) / result.size) * 100)
        
        if np.isnan(result).all():
            print(f"{band_name} is all NaNs after resampling — skipping this file.")
            skip_file = True
            break
        
        da = xr.DataArray(
            result,
            dims=("latitude", "longitude"),
            coords={"latitude": target_lats, "longitude": target_lons},
            name=band_name
        ).sel(
            longitude=slice(min_lon, max_lon),
            latitude=slice(min_lat, max_lat)
        )
        da_clipped = xr.DataArray(
            result,
            dims=("latitude", "longitude"),
            coords={"latitude": target_lats, "longitude": target_lons},
            name=band_name
        ).sel(
            longitude=slice(min_lon, max_lon),
            latitude=slice(min_lat, max_lat)
        )
        
        print(f"{band_name} valid pixel % (after clipping):", da_clipped.count().item() / da_clipped.size * 100)
        
        # Normalize per band (clip to 2–98% quantile range)
        if da.count() == 0:
            print(f"{band_name} is empty after clipping to bounding box — skipping.")
            skip_file = True
            break
        
        vmin = float(da.quantile(0.02))
        vmax = float(da.quantile(0.98))
        norm = np.clip((da - vmin) / (vmax - vmin), 0, 1)

        
        regridded_channels.append(norm)

    
    if skip_file == False: break

# ---------------------------
# Stack into RGB and Plot
# ---------------------------
rgb = np.stack(regridded_channels, axis=-1)
if np.isnan(rgb).all():
    print("All RGB image data is NaN — skipping plot.")
    exit()

# Normalize based on solar spectrum weights
solar_weights = np.array([0.8, 1.0, 1.1])
rgb_scaled = rgb / solar_weights
print("rgb_scaled min/max:", np.nanmin(rgb_scaled), np.nanmax(rgb_scaled))

# Avoid division by zero/NaN in normalization
rgb_max = np.nanmax(rgb_scaled)
if np.isnan(rgb_max) or rgb_max == 0:
    print("rgb_scaled max is invalid — skipping plot.")
else:
    rgb_scaled /= rgb_max
    rgb_gamma = np.clip(rgb_scaled ** (1/2.2), 0, 1)

    fig = plt.figure(figsize=(10, 7), dpi=100)
    ax = plt.axes(projection=ccrs.PlateCarree())
    ax.set_extent([min_lon, max_lon, min_lat, max_lat], crs=ccrs.PlateCarree())

    ax.imshow(
        rgb_gamma,
        origin='upper',
        extent=[min_lon, max_lon, min_lat, max_lat],
        transform=ccrs.PlateCarree()
    )

    ax.coastlines(resolution='10m', linewidth=0.5)
    ax.gridlines(draw_labels=False)
    ax.set_title("True Color Composite (Rrs 645/555/488)", loc="center")
    plt.show()
```
