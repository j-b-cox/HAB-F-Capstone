```python
import os
import datetime
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime, timedelta
import xarray as xr
from pyresample import geometry, kd_tree
import earthaccess
import cartopy.crs as ccrs
import cartopy.feature as cfeature
from cartopy.mpl.ticker import LongitudeFormatter, LatitudeFormatter

# Constants
PLOT_BBOX = (-83.5, 41.3, -82.45, 42.2)
BOX_HALF_DEG = 0.0225  # ~5km
TARGET_RESOLUTION = 0.0005  # ~1km grid

# Authenticate
auth = earthaccess.login(persist=True)

# Load station data
df = pd.read_csv("glrl-hab-data.csv", index_col=0)

def extract_datetime_from_modis_filename(path):
    filename = os.path.basename(path)
    try:
        timestamp_str = filename.split(".")[1]  # e.g., "20120514T173000"
        return datetime.strptime(timestamp_str, "%Y%m%dT%H%M%S")
    except Exception as e:
        print(f"[ERROR] Failed to extract datetime from {filename}: {e}")
        return None

for i, row in df.iterrows():
    if i < 1: continue

    station_lat = row['lat']
    station_lon = row['lon']
    timestamp = pd.to_datetime(row['timestamp'])
    print(f"\n[INFO] Processing row {i} | Station: {row['station_name']} | {timestamp}")

    mini_bbox = (
        station_lon - BOX_HALF_DEG,
        station_lat - BOX_HALF_DEG,
        station_lon + BOX_HALF_DEG,
        station_lat + BOX_HALF_DEG
    )

    start = (timestamp - timedelta(hours=48)).strftime('%Y-%m-%dT%H:%M:%SZ')
    end   = (timestamp + timedelta(hours=48)).strftime('%Y-%m-%dT%H:%M:%SZ')

    print(f"[INFO] Searching for granules in time range {start} to {end} within 5km box.")

    results = earthaccess.search_data(
        short_name="MODISA_L2_OC",
        temporal=(start, end),
        bounding_box=mini_bbox
    )

    if not results:
        print(f"[WARN] No granules found for station {row['station_name']}")
        continue

    paths = earthaccess.download(results, "granules")
    print(f"[INFO] {len(paths)} granules downloaded.")

    sorted_paths = sorted(paths, key=lambda p: abs((extract_datetime_from_modis_filename(p) or timestamp) - timestamp))

    found_valid = False
    for best_file in sorted_paths:
        print(f"[INFO] Trying granule: {os.path.basename(best_file)}")

        try:
            obs = xr.open_dataset(best_file, group="geophysical_data")
            nav = xr.open_dataset(best_file, group="navigation_data")
            nav = nav.set_coords(("longitude", "latitude"))
            dataset = xr.merge((obs, nav.coords))
        except Exception as e:
            print(f"[ERROR] Failed to load {best_file}: {e}")
            continue

        lon_min, lat_min, lon_max, lat_max = PLOT_BBOX
        margin = 0.1
        target_lons = np.arange(lon_min - margin, lon_max + margin, TARGET_RESOLUTION)
        target_lats = np.arange(lat_min - margin, lat_max + margin, TARGET_RESOLUTION)
        lon2d, lat2d = np.meshgrid(target_lons, target_lats)
        area_def = geometry.GridDefinition(lons=lon2d, lats=lat2d)

        lons = nav["longitude"].values.flatten()
        lats = nav["latitude"].values.flatten()

        bands = {
            "Rrs_667": "red",
            "Rrs_555": "green",
            "Rrs_488": "blue"
        }

        regridded_channels = []
        skip_file = False

        for band_name in bands:
            print(f"[INFO] Reprojecting band: {band_name}")
            data = dataset[band_name].values.flatten()
            valid = ~np.isnan(data) & ~np.isnan(lons) & ~np.isnan(lats)

            if not np.any(valid):
                print(f"[WARN] {band_name} has no valid data.")
                skip_file = True
                break

            swath_def = geometry.SwathDefinition(lons=lons[valid], lats=lats[valid])
            result = kd_tree.resample_nearest(
                swath_def, data[valid], area_def,
                radius_of_influence=5000,
                fill_value=np.nan
            )

            full_da = xr.DataArray(
                result,
                dims=("latitude", "longitude"),
                coords={"latitude": target_lats, "longitude": target_lons},
                name=band_name
            ).sel(
                longitude=slice(PLOT_BBOX[0], PLOT_BBOX[2]),
                latitude=slice(PLOT_BBOX[1], PLOT_BBOX[3])
            )

            test_da = full_da.sel(
                longitude=slice(station_lon - BOX_HALF_DEG, station_lon + BOX_HALF_DEG),
                latitude=slice(station_lat - BOX_HALF_DEG, station_lat + BOX_HALF_DEG)
            )

            valid_pct = test_da.count().item() / test_da.size * 100
            print(f"[INFO] {band_name} valid pixel % in 5km box: {valid_pct:.2f}%")

            if valid_pct < 2:
                print(f"[WARN] Insufficient data in 5km box for {band_name}")
                skip_file = True
                break

            regridded_channels.append(full_da)

        if skip_file:
            continue
        elif len(regridded_channels) != 3:
            print(f"[ERROR] Not all bands available.")
        else:
            found_valid = True
            break

    if not found_valid:
        print(f"[WARN] All granules failed for row {i}.")
        continue

    stacked = xr.concat(regridded_channels, dim="band")
    vmin = float(stacked.quantile(0.02))
    vmax = float(stacked.quantile(0.98))
    print(f"[INFO] Normalization range: vmin={vmin:.4f}, vmax={vmax:.4f}")

    regridded_channels = [
        np.clip((band - vmin) / (vmax - vmin), 0, 1)
        for band in regridded_channels
    ]

    rgb = np.stack(regridded_channels, axis=-1)
    rgb_gamma = np.clip(rgb ** (1/2.2), 0, 1)

    # Extract extent for imshow
    lat0 = float(regridded_channels[0].latitude.min())
    lat1 = float(regridded_channels[0].latitude.max())
    lon0 = float(regridded_channels[0].longitude.min())
    lon1 = float(regridded_channels[0].longitude.max())

    fig = plt.figure(figsize=(10, 7), dpi = 100)
    ax = plt.axes(projection=ccrs.PlateCarree())
    
    lon0 = float(regridded_channels[0].longitude.min())
    lon1 = float(regridded_channels[0].longitude.max())
    lat0 = float(regridded_channels[0].latitude.min())
    lat1 = float(regridded_channels[0].latitude.max())
    extent = [lon0, lon1, lat0, lat1]

    ax.imshow(
        rgb_gamma,
        origin='lower',
        extent=extent,
        transform=ccrs.PlateCarree()
    )

    ax.set_extent(extent, crs=ccrs.PlateCarree())
    ax.coastlines(resolution='10m', linewidth=0.5)
    ax.add_feature(cfeature.BORDERS, linewidth=0.5)
    ax.add_feature(cfeature.STATES, linewidth=0.5)
    ax.gridlines(draw_labels=True)
    ax.set_title(f"MODIS-Aqua True Color @ {row['station_name']} ({timestamp.date()})")

    #Draw red box around 5x5 km station area
    box = [
        [station_lon - BOX_HALF_DEG, station_lat - BOX_HALF_DEG],
        [station_lon - BOX_HALF_DEG, station_lat + BOX_HALF_DEG],
        [station_lon + BOX_HALF_DEG, station_lat + BOX_HALF_DEG],
        [station_lon + BOX_HALF_DEG, station_lat - BOX_HALF_DEG],
        [station_lon - BOX_HALF_DEG, station_lat - BOX_HALF_DEG]
    ]
    box_lons, box_lats = zip(*box)
    ax.plot(box_lons, box_lats, transform=ccrs.PlateCarree(), color='red', linewidth=2)

    ax.xaxis.set_major_formatter(LongitudeFormatter())
    ax.yaxis.set_major_formatter(LatitudeFormatter())

    plt.show()
    break
```

```python
fig = plt.figure(figsize=(10, 7), dpi=100)
ax = plt.axes(projection=ccrs.PlateCarree())

extent = [-83.5, -81.0, 41.3, 42.2]

ax.imshow(
    rgb_gamma,
    origin='lower',
    extent=extent,
    transform=ccrs.PlateCarree()
)

ax.set_extent(extent, crs=ccrs.PlateCarree())
ax.coastlines(resolution='10m', linewidth=0.5)
ax.add_feature(cfeature.BORDERS, linewidth=0.5)
ax.add_feature(cfeature.STATES, linewidth=0.5)
ax.add_feature(cfeature.LAND, facecolor='lightgray')
ax.gridlines(draw_labels=True)
ax.set_title(f"MODIS-Aqua True Color @ {row['station_name']} ({timestamp.date()})")

plt.show()
```
